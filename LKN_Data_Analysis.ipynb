```python
# Missing Values Analysis

# Define key columns and their associated platform suffixes
key_columns = {
    "name_deliveroo": "_deliveroo",
    "name_justeat": "_justeat",
    "name_ubereats": "_ubereats",
    "name_googleplaces": "_googleplaces"
}

# Define the fields you want to check
fields_of_interest = ['rating', 'reviews', 'food_types']

# Store results
custom_missing_report = []

for key_col, platform_suffix in key_columns.items():
    # Filter rows where key column is not null
    platform_df = df[df[key_col].notnull()]
    platform_len = len(platform_df)

    # Filter platform columns that match the fields of interest
    platform_columns = [
        col for col in df.columns
        if col.endswith(platform_suffix) and any(field in col for field in fields_of_interest)
    ]

    for col in platform_columns:
        if col in platform_df.columns:
            missing_count = platform_df[col].isnull().sum()
            missing_percentage = (missing_count / platform_len) * 100

            if missing_count > 0:
                custom_missing_report.append({
                    'Column': col,
                    'Missing Values': missing_count,
                    'Percentage (Non-Null Key Rows)': round(missing_percentage, 2),
                    'Reference Key': key_col,
                    'Non-Null Rows (Key)': platform_len
                })

# Create DataFrame
custom_missing_df = pd.DataFrame(custom_missing_report)

# Sort by highest missing percentage
custom_missing_df = custom_missing_df.sort_values(by='Percentage (Non-Null Key Rows)', ascending=False)

# Show top 10
print(custom_missing_df.head(10))
```

```python

#Summarise the dataset using key statistics (e.g., average ratings,total reviews per platform, number of unique brands).

# Define platforms and key columns
platforms = {
    'Deliveroo': 'name_deliveroo',
    'JustEat': 'name_justeat',
    'UberEats': 'name_ubereats',
    'GooglePlaces': 'name_googleplaces'
}

# Define the platform-specific columns you want to summarize
summary_metrics = {
    'Average Rating': {
        'Deliveroo': 'ratings_deliveroo',
        'JustEat': 'ratings_justeat',
        'UberEats': 'ratings_ubereats',
        'GooglePlaces': 'ratings_googleplaces'
    },
    'Total Reviews': {
        'Deliveroo': 'reviews_deliveroo',
        'JustEat': 'reviews_justeat',
        'UberEats': 'reviews_ubereats',
        'GooglePlaces': 'reviews_googleplaces'
    },
    'Unique Brands': {
        'Deliveroo': 'brand_deliveroo',
        'JustEat': 'brand_justeat',
        'UberEats': 'brand_ubereats',
        'GooglePlaces': None  # if no brand column for GooglePlaces
    }
}

# Create summary table
summary_data = []

for platform, key_col in platforms.items():
    platform_data = df[df[key_col].notnull()]
    
    avg_rating = platform_data[summary_metrics['Average Rating'][platform]].mean()
    total_reviews = platform_data[summary_metrics['Total Reviews'][platform]].sum()
    if summary_metrics['Unique Brands'][platform]:
        unique_brands = platform_data[summary_metrics['Unique Brands'][platform]].nunique()
    else:
        unique_brands = 'N/A'

    summary_data.append({
        'Platform': platform,
        'Avg Rating': round(avg_rating, 2),
        'Total Reviews': int(total_reviews),
        'Unique Brands': unique_brands
    })

# Convert to DataFrame
summary_df = pd.DataFrame(summary_data)
print(summary_df)
```

```python

#Identify the top 10 brands with the highest average rating. How reliable are these ratings based on the number of reviews?

# Filter data for non-null brand and rating columns per platform
df_deliveroo = df[df['name_deliveroo'].notna()][['brand_deliveroo', 'ratings_deliveroo', 'reviews_deliveroo']].copy()
df_deliveroo.columns = ['Brand', 'Rating', 'Reviews']
df_deliveroo['Platform'] = 'Deliveroo'

df_justeat = df[df['name_justeat'].notna()][['brand_justeat', 'ratings_justeat', 'reviews_justeat']].copy()
df_justeat.columns = ['Brand', 'Rating', 'Reviews']
df_justeat['Platform'] = 'JustEat'

df_ubereats = df[df['name_ubereats'].notna()][['brand_ubereats', 'ratings_ubereats', 'reviews_ubereats']].copy()
df_ubereats.columns = ['Brand', 'Rating', 'Reviews']
df_ubereats['Platform'] = 'UberEats'

# Combine all platforms
brand_data = pd.concat([df_deliveroo, df_justeat, df_ubereats], ignore_index=True)

# Remove rows with missing ratings or reviews
brand_data = brand_data.dropna(subset=['Brand', 'Rating', 'Reviews'])

# Group by Brand and Platform
grouped = brand_data.groupby(['Platform', 'Brand']).agg(
    Avg_Rating=('Rating', 'mean'),
    Total_Reviews=('Reviews', 'sum')
).reset_index()

# Optional: Filter out brands with too few reviews for reliability
filtered = grouped[grouped['Total_Reviews'] >= 50]

# Sort by Avg Rating
top_10_brands = filtered.sort_values(by='Avg_Rating', ascending=False).head(10)

# Optional: Add reliability indicator
def rating_reliability(reviews):
    if reviews >= 1000:
        return 'High'
    elif reviews >= 200:
        return 'Medium'
    else:
        return 'Low'

top_10_brands['Reliability'] = top_10_brands['Total_Reviews'].apply(rating_reliability)

top_10_brands
```

```python
# Are there brands that have high ratings but a low number of reviews? What does this suggest?

# Step 1: Unify brand column
df['brand'] = df[['brand_deliveroo', 'brand_justeat', 'brand_ubereats']].bfill(axis=1).iloc[:, 0]

# Step 2: Fill NaNs for reviews and ratings
review_cols = ['reviews_deliveroo', 'reviews_justeat', 'reviews_ubereats', 'reviews_googleplaces']
rating_cols = ['ratings_deliveroo', 'ratings_justeat', 'ratings_ubereats', 'ratings_googleplaces']
df[review_cols] = df[review_cols].fillna(0)
df[rating_cols] = df[rating_cols].fillna(np.nan)

# Step 3: Compute total reviews and average rating per row
df['total_reviews'] = df[review_cols].sum(axis=1)
df['average_rating'] = df[rating_cols].mean(axis=1)

# Step 4: Group by brand
brand_stats = df.groupby('brand').agg({
    'total_reviews': 'sum',
    'average_rating': 'mean'
}).reset_index()

# Step 5: Filter brands with high ratings but low review counts
suspiciously_high = brand_stats[(brand_stats['average_rating'] > 4.5) & (brand_stats['total_reviews'] < 100)]
# Step 6: Sort by average rating descending
suspiciously_high = suspiciously_high.sort_values(by='average_rating', ascending=False)


print(suspiciously_high)
```

```python
# Compare the average ratings across different delivery platforms. Which platform has the highest and lowest average rating?

# List of rating columns per platform
rating_cols = {
    'Deliveroo': 'ratings_deliveroo',
    'JustEat': 'ratings_justeat',
    'UberEats': 'ratings_ubereats',
    'GooglePlaces': 'ratings_googleplaces'
}

# Calculate average rating per platform
platform_avg_ratings = {
    platform: df[col].mean(skipna=True)
    for platform, col in rating_cols.items()
}

# Convert to DataFrame for nicer display
avg_rating_df = pd.DataFrame.from_dict(platform_avg_ratings, orient='index', columns=['Avg Rating']).reset_index()
avg_rating_df.rename(columns={'index': 'Platform'}, inplace=True)

# Identify the highest and lowest rated platforms
highest = avg_rating_df.loc[avg_rating_df['Avg Rating'].idxmax()]
lowest = avg_rating_df.loc[avg_rating_df['Avg Rating'].idxmin()]

# Print results
print("Average Rating by Platform:\n", avg_rating_df)
print("\nHighest Rated Platform:", highest['Platform'], "with", round(highest['Avg Rating'], 2))
print("Lowest Rated Platform:", lowest['Platform'], "with", round(lowest['Avg Rating'], 2))
```

```python
# Plotting the average ratings
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_rating_df, x='Platform', y='Avg Rating', palette='viridis')

# Add data labels on bars
for i, row in avg_rating_df.iterrows():
    plt.text(i, row['Avg Rating'] + 0.01, round(row['Avg Rating'], 2), ha='center', va='bottom', fontweight='bold')

plt.title('Average Ratings by Delivery Platform')
plt.ylim(0, 5.2)
plt.ylabel('Average Rating')
plt.xlabel('Platform')
plt.tight_layout()
plt.show()
```

```python
# Is there a correlation between the number of reviews and ratings for each platform? What insights can be drawn?

# Ensure review columns are filled with 0, and ratings can stay NaN
review_cols = ['reviews_deliveroo', 'reviews_justeat', 'reviews_ubereats', 'reviews_googleplaces']
rating_cols = ['ratings_deliveroo', 'ratings_justeat', 'ratings_ubereats', 'ratings_googleplaces']
df[review_cols] = df[review_cols].fillna(0)

# Subset only the relevant columns for correlation
correlation_df = df[review_cols + rating_cols]

# Compute correlation matrix
correlation_matrix = correlation_df.corr()

# Display correlation between reviews and ratings per platform
for platform in ['deliveroo', 'justeat', 'ubereats', 'googleplaces']:
    corr = correlation_matrix.loc[f'reviews_{platform}', f'ratings_{platform}']
    print(f"{platform.capitalize()} - Correlation between reviews and ratings: {round(corr, 3)}")
```

```python
# Identify the food types with the highest and lowest average ratings. What factors could contribute to these variations?

# Step 1: Unify food type column (based on your naming)
df['food_type'] = df[['food_types_deliveroo', 'food_types_justeat', 'food_types_ubereats']].bfill(axis=1).iloc[:, 0]

# Step 2: Handle missing/zero ratings
rating_cols = ['ratings_deliveroo', 'ratings_justeat', 'ratings_ubereats']
df[rating_cols] = df[rating_cols].replace(0, np.nan)

# Step 3: Compute average rating across platforms
df['average_rating'] = df[rating_cols].mean(axis=1)

# Step 4: Group by food type to get average rating
food_type_stats = df.groupby('food_type')['average_rating'].mean().reset_index()

# Step 5: Sort and display top and bottom food types
top_food_types = food_type_stats.sort_values(by='average_rating', ascending=False).head(5)
bottom_food_types = food_type_stats.sort_values(by='average_rating').head(5)

# Output
print("Top 5 Food Types by Average Rating:\n", top_food_types)
print("\nBottom 5 Food Types by Average Rating:\n", bottom_food_types)
```

```python
# Are there food types where a specific delivery platform performs significantly better or worse?

# Step 1: Unify brand and food_type columns
df['brand'] = df[['brand_deliveroo', 'brand_justeat', 'brand_ubereats']].bfill(axis=1).iloc[:, 0]
df['food_type'] = df[['food_types_deliveroo', 'food_types_justeat', 'food_types_ubereats']].bfill(axis=1).iloc[:, 0]

# Step 2: Create a long-form dataframe with platform-specific ratings
platform_ratings = pd.DataFrame({
    'brand': df['brand'],
    'food_type': df['food_type'],
    'Deliveroo': df['ratings_deliveroo'],
    'JustEat': df['ratings_justeat'],
    'UberEats': df['ratings_ubereats']
})

# Step 3: Melt to long format
melted = platform_ratings.melt(
    id_vars=['brand', 'food_type'],
    value_vars=['Deliveroo', 'JustEat', 'UberEats'],
    var_name='platform',
    value_name='rating'
)

# Step 4: Drop null ratings
melted = melted.dropna(subset=['rating'])

# Step 5: Group by food_type and platform, calculate average rating
avg_ratings = melted.groupby(['food_type', 'platform'])['rating'].mean().reset_index()

# Step 6: Pivot to compare platforms side by side
pivot_ratings = avg_ratings.pivot(index='food_type', columns='platform', values='rating')

# Step 7: Add a column for max - min rating to find high variation
pivot_ratings['rating_range'] = pivot_ratings.max(axis=1) - pivot_ratings.min(axis=1)

# Step 8: Filter food types with significant platform rating difference (> 1.0)
significant_diff = pivot_ratings[pivot_ratings['rating_range'] > 1.0].sort_values('rating_range', ascending=False)

# View the results
print(significant_diff.head(10))
```
